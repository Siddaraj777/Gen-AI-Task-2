```python
# ==========================================
# LANGCHAIN ALL-IN-ONE ASSIGNMENT CODE
# FULL PDF PATH VERSION
# ==========================================

from langchain_openai import ChatOpenAI, OpenAIEmbeddings
from langchain_core.prompts import PromptTemplate
from langchain_classic.memory import ConversationBufferMemory
from langchain_classic.chains import LLMChain, RetrievalQA
from langchain_classic.agents import Tool, initialize_agent
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
import os

# ==========================================
# API KEY
# ==========================================
from dotenv import load_dotenv
import os

load_dotenv()

# ==========================================
# 1) BASIC LLM CALL
# ==========================================
print("===== BASIC LLM CALL =====")

llm = ChatOpenAI(
    model="gpt-4.1-mini",
    temperature=0.7
)

response = llm.invoke("What is LangChain?")
print(response.content)

# ==========================================
# 2) PROMPT TEMPLATE
# ==========================================
print("\n===== PROMPT TEMPLATE =====")

prompt = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in 3 simple points"
)

print(prompt.format(topic="Machine Learning"))

# ==========================================
# 3) SIMPLE CHAIN
# ==========================================
print("\n===== SIMPLE CHAIN =====")

chain = LLMChain(llm=llm, prompt=prompt)
print(chain.run("LangChain"))

# ==========================================
# 4) MEMORY
# ==========================================
print("\n===== MEMORY =====")

memory = ConversationBufferMemory()
memory.save_context(
    {"input": "My name is Siddaraj"},
    {"output": "Nice to meet you"}
)
print(memory.load_memory_variables({}))

# ==========================================
# 5) AGENT + TOOL
# ==========================================
print("\n===== AGENT =====")

def square_tool(x):
    return str(int(x) ** 2)

tools = [
    Tool(
        name="SquareTool",
        func=square_tool,
        description="Returns square of a number"
    )
]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent="zero-shot-react-description"
)

print(agent.run("What is square of 12?"))

# ==========================================
# 6) PDF LOADER + FAISS + QA
# ==========================================
print("\n===== PDF LOADER =====")

# PASTE YOUR COPIED FULL PDF PATH HERE
pdf_path = r"C:\Users\dell\Downloads\LangChain Deep Technical Blog (Blog Writing Task).pdf"

try:
    loader = PyPDFLoader(pdf_path)
    documents = loader.load()
    print(f"Loaded {len(documents)} pages from PDF")

    print("\n===== VECTOR STORE =====")
    embeddings = OpenAIEmbeddings()
    vectorstore = FAISS.from_documents(documents, embeddings)
    print("FAISS created successfully")

    print("\n===== RETRIEVAL QA =====")
    qa_chain = RetrievalQA.from_chain_type(
        llm=llm,
        retriever=vectorstore.as_retriever()
    )

    answer = qa_chain.run("Give a summary of the main points in this document")
    print(answer)

except Exception as e:
    print("PDF loading failed:", e)

# ==========================================
# FINAL MESSAGE
# ==========================================
print("\n===== ALL TASKS COMPLETED SUCCESSFULLY =====")
```
